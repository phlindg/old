```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import scipy.ndimage
```

```python
csv_dir = "C:/Users/Phili/Desktop/fond/data/"
symbol ="SAND"
data = pd.read_csv(csv_dir+symbol+".csv", sep=";", header=0, index_col=0, parse_dates = True).iloc[::-1]
close = data["Closing price"].values
plt.plot(data["Closing price"])
```

```python
rets = np.diff(np.log(close))
plt.plot(rets)
```

```python
cdf = scipy.stats.norm.cdf(rets)
plt.plot(cdf)
```

```python
sigma = np.std(rets)
filt = scipy.ndimage.gaussian_filter(cdf, sigma=sigma, order=0)
plt.plot(filt)
```

```python
pdf = np.gradient(filt)
h = plt.hist(pdf, bins=100)
```
